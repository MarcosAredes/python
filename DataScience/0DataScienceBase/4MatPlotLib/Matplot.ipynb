```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
#Titulo
plt.title('Valores da gasolina de 1990 até 2008')

#Paises e legendas e cores
list_pais = ['USA', 'UK', 'Italy', 'Mexico', 'Japan']
for pais in gas:
    if pais in list_pais: 
        plt.plot(gas['Year'], gas[pais], label=[pais], marker='.')
#Eixos
plt.xticks(gas['Year'] [:: 2])

plt.xlabel('Anos')
plt.ylabel('Valor em Dolar US$')

#Imprimindo grafico e legenda
plt.legend()
plt.savefig('País.png')
plt.show()
```
